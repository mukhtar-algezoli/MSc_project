```python
from tqdm import tqdm
from pathlib import Path
import shutil
from sklearn.decomposition import PCA
import numpy as np

layer_num = 12
model = "Hubert_ASR"
nested_dir = "/content/drive/MyDrive/Grad project/models_hidden_states/" + model +"/"
input_file = nested_dir + "/test-clean"
input_files = list(Path(input_file).rglob("*.npy")) # CPC representations directory
```

```python
import os
# dir_path = "/content/drive/MyDrive/Grad project/models_hidden_states/Wav2Vec2/layers_avg/layer_" + str(layer_num)
output_file = nested_dir + "test-clean-layer"+ str(layer_num) + "_onefile"
print(output_file)
os.makedirs(output_file, exist_ok=True)
```

```python
counter = 0
for file in tqdm(input_files):
  loaded_representations = np.load(file)
  # if loaded_representations.size > 1:
  #   pass
  # else:
  #   print("hello")
  # print(loaded_representations.shape)
  layer_6_reps = loaded_representations[layer_num].squeeze()
  if counter <= 0:
    print(layer_6_reps.shape)
    counter += 1
  # layer_6_reps = loaded_representations
  path = output_file + "/" + str(file).split("/")[-1]
  # print(path)
  np.save(path, layer_6_reps)
```
